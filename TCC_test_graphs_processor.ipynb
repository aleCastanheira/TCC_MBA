```python
# Import necessary libraries
import os
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
from google.colab import drive
drive.mount('drive')
```

```python
def collage(directory):
    images = [i for i in os.listdir(directory) if i.endswith(".jpg") or i.endswith(".jpeg") or i.endswith(".png")]

    index = 0
    array = []
    for filename in images:
        # 3 linhas, 2 colunas e a posição do índice
        plt.subplot(3, 2, index+1)

        #leio a imagem usando opencv
        img = cv2.imread(directory + '/' + filename)

        # convertendo de BGR para RGB
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        array.append(img)

    hor_img1 = np.hstack((array[0], array[1]))
    hor_img2 = np.hstack((array[2], array[3]))
    # hor_img3 = np.hstack((array[4], array[5]))

    #empilhando as imagens verticalmente
    temp_img = np.vstack((hor_img1, hor_img2))

    #exibo a imagem
    plt.figure(figsize=(10,20))
    plt.imshow(temp_img)
    plt.axis('off')
    plt.savefig(directory+"/collage/collage.png", pad_inches=0, bbox_inches='tight')
    plt.show()

def execute():
    directory = "/content/drive/MyDrive/TCC/imgs/TEST"
    number_of_tests = 6

    # Get list of images
    for test_n in range(1, number_of_tests + 1):
        collage(directory + str(test_n))
```

```python
execute()
```
